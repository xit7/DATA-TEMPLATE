# Expected Values

> **Expected values allow us to compare uncertain outcomes**. The expected value of a random outcome is the long-run average payoff you’d get if you could repeat the same uncertain situation many times under the same probabilities.

$EV = Σ(Probability × Outcome)$

Mathematically **discrete**: $\mathrm{EV}(X)=\sum_i p_i \, x_i$, **continuous**: $\mathrm{EV}(X)=\int x\,f(x)\,dx$

> So the $\text{Expected Value} = \text{weighted average of ALL possible outcomes}$

Note: The SUM across all possible outcomes is already the weighted average, so no need to divide by number of outcomes as you would do for a normal average.

**Simple Average**
**Average** of [10, 20, 30] = (10 + 20 + 30) / 3 = 20

**Weighted Average (when outcomes have different probabilities)**
* Outcome A: value 10, probability 50%
* Outcome B: value 20, probability 30%
* Outcome C: value 30, probability 20%

**Weighted Average** = (0.5 × 10) + (0.3 × 20) + (0.2 × 30) = 17
The probabilities already "do the dividing" — they sum to 100%, so multiplying and summing gives you the average.

## Example: Coin Flipping
How will flipping a coin turn out over time? Suppose you gain 10$ on heads and -10$ on tails. The expected value caluclation for this:

$E_{(price)} = p_{(tails)}\times{(-10)} + p_{(heads)}\times{10}$

If our coin is fair - so that each side has 50% probability of showing - the expected value is $0.5\times-10+0.5\times10=0$.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(1005)

# Number of trials
N=10000

total_earnings = pd.DataFrame(index= np.arange(N), columns= ['earnings'])

for i in range(N):
    draw = np.random.rand() # random number each draw

    if draw >= 0.5:
        curr_earn = 10 # win
    else:
        curr_earn = -10 # loose
    total_earnings.loc[i, 'earnings'] = curr_earn # store value

total_earnings['cumulative_earnings'] = total_earnings.earnings.cumsum() # cummulative sum
total_earnings['trials'] = np.arange(1, N+1) # simple counter
total_earnings['avg_earnings'] = total_earnings.cumulative_earnings / total_earnings.trials # average
```

## Example: Should We Build a Premium Feature?
* Business Context: SaaS company with 10,000 users. Considering building "Advanced Analytics" feature.
* Cost: $50,000 development
* Outcomes depend on: user adoption & churn impact

### Mathematical Concept

**Option A: Build Feature**
* Best case (30%): 500 users upgrade to premium (+$20/mo), reduce churn by 2%
* Base case (50%): 300 users upgrade, reduce churn by 1%
* Worst case (20%): 100 users upgrade, no churn impact

**Option B: Don't Build**
* Status quo: $0 cost, $0 benefit
* Calculate 12-month EV for each scenario.

### Implementation Steps
1. **Data Preparation**: Define scenarios with probabilities and financial impacts

2. **Calculate Expected Values**: Compute weighted average of all outcomes

3. **Compare & Decide**: Choose option with highest EV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# === 1. BUSINESS DATA ===
current_users = 10000
avg_revenue_per_user = 10  # $/month
dev_cost = 50000
premium_price = 20  # $/month additional
timeframe = 12  # months

# === 2. DEFINE SCENARIOS ===
scenarios = {
    'Best Case': {
        'probability': 0.30,
        'premium_adopters': 500,
        'churn_reduction': 0.02  # 2% absolute reduction
    },
    'Base Case': {
        'probability': 0.50,
        'premium_adopters': 300,
        'churn_reduction': 0.01
    },
    'Worst Case': {
        'probability': 0.20,
        'premium_adopters': 100,
        'churn_reduction': 0.00
    }
}

# === 3. CALCULATE OUTCOMES FOR EACH SCENARIO ===
results = []

for scenario_name, params in scenarios.items():
    prob = params['probability']
    adopters = params['premium_adopters']
    churn_reduction = params['churn_reduction']
    
    # Revenue from premium upgrades
    premium_revenue = adopters * premium_price * timeframe
    
    # Revenue from reduced churn (users retained * avg revenue)
    retained_users = current_users * churn_reduction
    churn_revenue = retained_users * avg_revenue_per_user * timeframe
    
    # Total benefit
    total_benefit = premium_revenue + churn_revenue
    
    # Net outcome (benefit - cost)
    net_outcome = total_benefit - dev_cost
    
    # Weighted outcome (for EV calculation)
    weighted_outcome = prob * net_outcome
    
    results.append({
        'Scenario': scenario_name,
        'Probability': prob,
        'Premium Revenue': premium_revenue,
        'Churn Revenue': churn_revenue,
        'Total Benefit': total_benefit,
        'Net Outcome': net_outcome,
        'Weighted Outcome': weighted_outcome
    })

# === 4. CREATE DATAFRAME ===
df = pd.DataFrame(results)
print("Calculated Scenario Outcomes:")
df
```

```python
# === 5. CALCULATE EXPECTED VALUE ===
ev_build_feature = df['Weighted Outcome'].sum()
ev_do_nothing = 0  # Status quo

print("=" * 60)
print("EXPECTED VALUE ANALYSIS: PREMIUM FEATURE")
print("=" * 60)
print(f"\nDevelopment Cost: ${dev_cost:,}")
print(f"Analysis Period: {timeframe} months\n")

print(df.to_string(index=False))

print("\n" + "=" * 60)
print(f"Expected Value (Build Feature): ${ev_build_feature:,.2f}")
print(f"Expected Value (Do Nothing):    ${ev_do_nothing:,.2f}")
print("=" * 60)
```

```python
# === 6. DECISION ===
if ev_build_feature > ev_do_nothing:
    decision = "BUILD THE FEATURE"
    advantage = ev_build_feature - ev_do_nothing
    print(f"\n✓ RECOMMENDATION: {decision}")
    print(f"  Expected advantage: ${advantage:,.2f} over 12 months")
else:
    decision = "DON'T BUILD"
    print(f"\n✗ RECOMMENDATION: {decision}")

# === 7. VISUALIZATION ===
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1: Scenario outcomes
scenarios_list = df['Scenario']
net_outcomes = df['Net Outcome']
colors = ['#2ecc71', '#3498db', '#e74c3c']

ax1.bar(scenarios_list, net_outcomes, color=colors, alpha=0.7, edgecolor='black')
ax1.axhline(y=0, color='black', linestyle='--', linewidth=1)
ax1.axhline(y=ev_build_feature, color='orange', linestyle='-', linewidth=2, label=f'EV = ${ev_build_feature:,.0f}')
ax1.set_ylabel('Net Outcome ($)')
ax1.set_title('Scenario Outcomes (12 months)')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

# Format y-axis
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Chart 2: Weighted contribution to EV
weighted = df['Weighted Outcome']
ax2.bar(scenarios_list, weighted, color=colors, alpha=0.7, edgecolor='black')
ax2.axhline(y=0, color='black', linestyle='--', linewidth=1)
ax2.set_ylabel('Weighted Contribution ($)')
ax2.set_title('Contribution to Expected Value')
ax2.grid(axis='y', alpha=0.3)

# Format y-axis
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

plt.tight_layout()
plt.show()

# === 8. SENSITIVITY ANALYSIS ===
print("\n" + "=" * 60)
print("SENSITIVITY: What if Base Case probability changes?")
print("=" * 60)

base_probs = np.linspace(0.3, 0.7, 5)
for p_base in base_probs:
    # Redistribute probabilities
    p_best = (1 - p_base) * 0.4
    p_worst = (1 - p_base) * 0.6
    
    ev_adjusted = (
        p_best * (df.loc[0, 'Net Outcome']) +
        p_base * (df.loc[1, 'Net Outcome']) +
        p_worst * (df.loc[2, 'Net Outcome'])
    )
    
    print(f"Base Case Prob = {p_base:.1%} → EV = ${ev_adjusted:,.2f}")
```

## How to come up with estimates on probabilities and outcomes?
In the example above, the probabilities and outcomes were assumed. In real life, you can use:
* Historical data analysis
* Expert judgment
* Market research
* Sensitivity analysis
* etc.
* Combine multiple methods for robust estimates.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# === HISTORICAL DATA ===
# Past premium features we've launched
historical_features = pd.DataFrame({
    'Feature': ['Export PDF', 'Custom Branding', 'API Access', 'Priority Support', 
                'Advanced Filters', 'Team Collaboration', 'White Label'],
    'Launch_Date': ['2023-01', '2023-04', '2023-07', '2023-10', 
                    '2024-01', '2024-04', '2024-07'],
    'Users_at_Launch': [8000, 8500, 8800, 9200, 9500, 9800, 10000],
    'Premium_Adopters': [280, 420, 310, 150, 330, 450, 290],
    'Price_Point': [15, 25, 20, 15, 20, 25, 20]
})
```

```python
historical_features
```

```python
# Calculate adoption rate (%)
historical_features['Adoption_Rate_%'] = (
    historical_features['Premium_Adopters'] / historical_features['Users_at_Launch'] * 100
)

print("=" * 70)
print("HISTORICAL PREMIUM FEATURE ADOPTION")
print("=" * 70)
print(historical_features[['Feature', 'Premium_Adopters', 'Adoption_Rate_%']].to_string(index=False))
```

```python

# === STATISTICAL ANALYSIS ===
adoption_rates = historical_features['Adoption_Rate_%']

mean_rate = adoption_rates.mean()
std_rate = adoption_rates.std()
median_rate = adoption_rates.median()

print("\n" + "=" * 70)
print("STATISTICAL SUMMARY")
print("=" * 70)
print(f"Mean Adoption Rate:     {mean_rate:.2f}%")
print(f"Median Adoption Rate:   {median_rate:.2f}%")
print(f"Std Deviation:          {std_rate:.2f}%")
print(f"Min:                    {adoption_rates.min():.2f}%")
print(f"Max:                    {adoption_rates.max():.2f}%")
```

```python
# Plot adoption rate histogram
plt.figure(figsize=(8, 5))
plt.hist(adoption_rates, bins=5, color='#3498db', alpha=0.7, edgecolor='black')
plt.axvline(mean_rate, color='red', linestyle='--', label=f'Mean: {mean_rate:.2f}%')
plt.axvline(median_rate, color='green', linestyle='-', label=f'Median: {median_rate:.2f}%')
plt.title('Histogram of Premium Feature Adoption Rates')
plt.xlabel('Adoption Rate (%)')
plt.ylabel('Number of Features')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python

# === METHOD 1: PERCENTILE-BASED SCENARIOS ===
print("\n" + "=" * 70)
print("METHOD 1: PERCENTILE-BASED PROBABILITY ESTIMATION")
print("=" * 70)

# Define scenarios based on percentiles
percentile_25 = np.percentile(adoption_rates, 25)  # Worst case
percentile_50 = np.percentile(adoption_rates, 50)  # Base case
percentile_75 = np.percentile(adoption_rates, 75)  # Best case

print(f"\nWorst Case (25th percentile): {percentile_25:.2f}% adoption")
print(f"Base Case (50th percentile):  {percentile_50:.2f}% adoption")
print(f"Best Case (75th percentile):  {percentile_75:.2f}% adoption")

# Convert to number of users (assuming 10,000 user base)
current_users = 10000
worst_adopters = int(current_users * percentile_25 / 100)
base_adopters = int(current_users * percentile_50 / 100)
best_adopters = int(current_users * percentile_75 / 100)

print(f"\nTranslated to user counts:")
print(f"Worst Case: {worst_adopters} users")
print(f"Base Case:  {base_adopters} users")
print(f"Best Case:  {best_adopters} users")

# Assign probabilities (typical distribution: focus on middle)
prob_worst = 0.20
prob_base = 0.60
prob_best = 0.20

print(f"\nAssigned Probabilities:")
print(f"Worst Case: {prob_worst:.0%}")
print(f"Base Case:  {prob_base:.0%}")
print(f"Best Case:  {prob_best:.0%}")
```

```python
# === METHOD 2: SIMILAR FEATURE CLUSTERING ===
print("\n" + "=" * 70)
print("METHOD 2: FILTER BY SIMILAR FEATURES")
print("=" * 70)

# New feature: "Advanced Analytics" at $20 price point
target_price = 20

# Filter to similar price point (±$5)
similar_features = historical_features[
    (historical_features['Price_Point'] >= target_price - 5) &
    (historical_features['Price_Point'] <= target_price + 5)
]

print(f"\nFeatures with similar price point (${target_price} ± $5):")
print(similar_features[['Feature', 'Price_Point', 'Adoption_Rate_%']].to_string(index=False))

similar_rates = similar_features['Adoption_Rate_%']
similar_mean = similar_rates.mean()
similar_std = similar_rates.std()

print(f"\nMean adoption (similar features): {similar_mean:.2f}%")
print(f"Std deviation (similar features): {similar_std:.2f}%")

# Define scenarios using ±1 standard deviation
worst_rate_2 = max(0, similar_mean - similar_std)
base_rate_2 = similar_mean
best_rate_2 = similar_mean + similar_std

print(f"\nScenario Estimates:")
print(f"Worst Case (mean - 1σ): {worst_rate_2:.2f}% → {int(current_users * worst_rate_2/100)} users")
print(f"Base Case (mean):       {base_rate_2:.2f}% → {int(current_users * base_rate_2/100)} users")
print(f"Best Case (mean + 1σ):  {best_rate_2:.2f}% → {int(current_users * best_rate_2/100)} users")

# Probabilities based on normal distribution (68-95-99.7 rule)
# Within 1σ: ~68%, so tails get ~16% each, middle gets ~68%
prob_worst_2 = 0.16
prob_base_2 = 0.68
prob_best_2 = 0.16

print(f"\nProbabilities (based on normal distribution):")
print(f"Worst Case: {prob_worst_2:.0%}")
print(f"Base Case:  {prob_base_2:.0%}")
print(f"Best Case:  {prob_best_2:.0%}")
```

```python
# === VISUALIZATION ===
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1: Historical adoption distribution
ax1.hist(adoption_rates, bins=5, color='#3498db', alpha=0.7, edgecolor='black')
ax1.axvline(mean_rate, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_rate:.1f}%')
ax1.axvline(percentile_25, color='orange', linestyle=':', linewidth=2, label=f'25th %ile: {percentile_25:.1f}%')
ax1.axvline(percentile_75, color='green', linestyle=':', linewidth=2, label=f'75th %ile: {percentile_75:.1f}%')
ax1.set_xlabel('Adoption Rate (%)')
ax1.set_ylabel('Frequency')
ax1.set_title('Historical Feature Adoption Distribution')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

# Chart 2: Scenario comparison
methods = ['Method 1\n(Percentiles)', 'Method 2\n(Similar Features)']
worst_cases = [worst_adopters, int(current_users * worst_rate_2/100)]
base_cases = [base_adopters, int(current_users * base_rate_2/100)]
best_cases = [best_adopters, int(current_users * best_rate_2/100)]

x = np.arange(len(methods))
width = 0.25

ax2.bar(x - width, worst_cases, width, label='Worst Case', color='#e74c3c', alpha=0.7)
ax2.bar(x, base_cases, width, label='Base Case', color='#3498db', alpha=0.7)
ax2.bar(x + width, best_cases, width, label='Best Case', color='#2ecc71', alpha=0.7)

ax2.set_ylabel('Estimated Adopters')
ax2.set_title('Comparison: Estimation Methods')
ax2.set_xticks(x)
ax2.set_xticklabels(methods)
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# === KEY TAKEAWAYS ===
print("\n" + "=" * 70)
print("KEY TAKEAWAYS")
print("=" * 70)
print("""
1. PERCENTILE METHOD: Use when you want conservative/moderate/optimistic splits
   - Based on actual historical distribution
   - Easy to explain to stakeholders

2. SIMILAR FEATURES METHOD: Use when you have comparable past launches
   - More accurate for similar price/complexity
   - Accounts for feature-specific factors

3. PROBABILITY ASSIGNMENT:
   - Percentile: Often use 20-60-20 or 25-50-25 split
   - Normal distribution: Use 16-68-16 (based on ±1σ rule)
   - Adjust based on confidence and risk tolerance

4. ALWAYS validate assumptions with domain experts!
""")
```
