
# EXAMPLE: Analytics One-Pager — KR: Increase Invite Acceptance Rate

## 1) KR Overview
- **KR Statement:** Increase ACPC invite acceptance rate to 50% by end of Q2 FY26.
- **Target / Success Measure:** Acceptance Rate = 50% (Impact metric); Balancing metric = Invite Quality Score ≥ 80.

## 2) Definitions
- **Acceptance:** A user clicks the invite and completes onboarding within 7 days.
- **Active user:** Signed in ≥1 session/week in product.
- **Scope:** In: email + in-app invites; Out: internal test users, partner sandbox tenants.

## 3) Baseline
- **Current Value (as of Dec 31, FY25; window=Q4 FY25):** 30% (example).
- **Data Source(s):** ACPC Usage Dashboard; lineage: `dw.marketing.invites` + `dw.product.onboarding`.
- **Data Freshness / SLA:** Daily; 95% on-time; late load triggers backup query.
- **Quality Note(s):** Excludes historical spam-filtered campaigns (Sept FY25).

## 4) Measurement Details
- **Metric Formula:** `Acceptance Rate = accepted_invites / total_invites_sent` filtered to eligible audience; time grain = week.
- **Dimensions & Cuts:** Geo (EMEA/AMER/APAC), segment (new vs. existing), channel (email/in-app).
- **Instrumentation Requirements:** Ensure `invite_accept` event with `campaign_id`, `tenant_id`, `channel`, `timestamp`; add `onboarding_complete` attribute.
- **Technical Implementation:** Warehouse: `dw.marketing.invites (sent_ts, campaign_id, user_id)`, `dw.product.events (event='invite_accept')`; Validation: join completeness ≥ 98%; pipeline `etl_invite_accept_v2`.
- **Dashboards:** ACPC OKR Dashboard (Team view), Executive Progress Board (Impact metrics).
- **Governance Alignment:** Metric directly informs invite targeting decisions; avoid vanity metrics (e.g., opens) unless tied to conversion path.

## 5) Risks & Mitigations
- **Data:** Ingestion delay during releases → mitigation: grace period + backup aggregation.
- **Methodology:** Definition change for “acceptance” → mitigation: versioned definition + backfill note.
- **Operational:** Dependency on lifecycle team to improve targeting → mitigation: weekly experiment cadence.

## 6) Roles & Ownership
- **DS Owner / Reviewer:** <Your name>; ensures formula accuracy & insights.
- **DE Owner:** <Pipeline owner>; maintains ETL quality & SLAs.
- **PM / KR Stakeholders:** Lifecycle PM; Growth team.
- **Governance Approver(s):** Analytics governance lead.

## 7) Status & Actions
- **Current Progress:** 32% (+2pp vs. baseline); leading indicators trending up (invite CTR +0.8pp).
- **Next Steps:** Experiment with subject lines A/B; tighten eligibility; owners + ETAs listed.
- **Review Cadence:** Mid‑quarter check‑in; QBR update in Quantive.

## 8) Change Log
- v0.2 (Jan 12, FY26): Added “7‑day onboarding” to acceptance definition; governance sign‑off logged.

## 9) Privacy & Compliance
- PII minimized; hashed user IDs in analysis; consent verified per campaign policy.
