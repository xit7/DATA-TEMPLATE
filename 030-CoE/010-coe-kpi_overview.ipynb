# Center of Excellence — KPI Overview

> A reference sheet for KPIs tracked within the CoE program. Each metric includes its definition, what it signals, and how it is calculated.

## What is a Center of Excellence?
A CoE is an internal team or function that drives standards, best practices, and capability-building across an organization. Measuring a CoE means tracking **both the output it produces** (adoption, quality, speed) and **the organizational impact** it creates (efficiency gains, consistency, maturity uplift).

### How to read this sheet
* **What it tells you** — the business question the metric answers.
* **Formula** — how to compute it.
* A metric is only useful if it changes behavior. If you cannot act on it, it is a vanity metric.

# Map of CoE KPIs

## Terminology
* **Lagging indicator** — measures outcomes of past activity (e.g. adoption rate at end of quarter).
* **Leading indicator** — predicts future outcomes (e.g. number of teams currently onboarding).
* **Health metric** — signals whether the CoE function itself is operating well.
* **Impact metric** — signals value delivered to the wider organization.

| KPI | What it Tells You | Formula / Measurement |

| **MAU** | Active user base size in trailing 28 days — the baseline volume metric. | $\text{MAU} = \lvert\{u : u \text{ active in trailing 28d}\}\rvert$ |
| **WAU** | Active user base size in trailing 7 days — more sensitive to recent shifts than MAU. | $\text{WAU} = \lvert\{u : u \text{ active in trailing 7d}\}\rvert$ |

### Collab MAU

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Leading indicator |
| **Cadence** | Monthly (rolling 28-day window, refreshed continuously) |
| **Formula** | $\text{Collab MAU} = \lvert\{u \in U : \exists\, e \in \text{Events}_{[-28d,\, 0]} \mid \text{actor}(e) = u \;\wedge\; \text{is\_shared}\bigl(\text{target}(e)\bigr)\}\rvert$ |
| **Target / Threshold** | _(to be defined — e.g. ↑ MoM, or ≥ X% of eligible users)_ |
| **Data Source** | _(collaboration platform event log)_ |

**What it tells you:** The count of distinct users who touched at least one shared platform artifact — workspace, dashboard, template, or similar — in the past 28 days.

**Why it matters for a CoE:** Collab MAU is the top-of-funnel adoption signal for shared infrastructure. If users are not engaging with shared constructs, CoE-published standards and resources are going unused — regardless of how many were published. A rising Collab MAU indicates the platform is becoming part of day-to-day workflows; a flat or declining number signals friction, low awareness, or poor relevance of the shared content.

**Key distinctions:**
- Counts **unique users**, not events — one user opening five shared dashboards counts as 1.
- "Shared platform construct" means any artifact with a multi-user scope: shared workspaces, published dashboards, team templates, shared notebooks, etc. Personal/private artifacts do not count.
- The **rolling 28-day window** (vs. a calendar month) smooths weekend/holiday dips and gives a stable daily-refreshable number.

**Example:** 420 users opened or edited at least one shared workspace or template between June 3 and June 30 → Collab MAU = **420**.

**Related metrics to watch alongside it:**
- Collab MAU as % of total eligible users → normalizes for team size growth.
- New vs. returning Collab MAU users → distinguishes acquisition from retention of collaborative behavior.

### M1RR — Month-1 Return Rate

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Leading indicator |
| **Cadence** | Monthly (computed once month N+1 is complete) |
| **Formula** | $\text{M1RR}(N) = \dfrac{\lvert\text{New users in month } N \text{ who return in month } N{+}1\rvert}{\lvert\text{New users acquired in month } N\rvert}$ |
| **Target / Threshold** | _(to be defined — e.g. ≥ 40%, or ↑ MoM)_ |
| **Data Source** | _(collaboration platform event log, user acquisition log)_ |

**What it tells you:** Of every cohort of brand-new users the platform acquires in a given month, what fraction comes back and does anything at all the very next month.

**Why it matters for a CoE:** M1RR is the earliest leading signal of whether new users find enough value to return. A high M1RR means onboarding is working and the platform delivers immediate utility. A low M1RR — even with strong new-user acquisition — means the CoE is filling a leaky bucket: users arrive but don't stick, and Collab MAU will eventually stagnate or decline as the churn outpaces acquisition.

**Key distinctions:**
- This is a **cohort metric**: each month N has its own M1RR, computed once month N+1 closes. You cannot update a past M1RR score.
- "Return" means any engagement in month N+1 — not necessarily the same action that brought the user in. The bar is presence, not depth.
- M1RR ≠ overall retention rate. It measures only the **first cohort transition** (month 0 → month 1). Use longer-horizon metrics (M3RR, M6RR) for sustained retention.

**Example:** In April, 200 new users joined the platform. By end of May, 90 of those 200 had returned at least once → M1RR(April) = **90 ÷ 200 = 45%**.

**Related metrics to watch alongside it:**
- **Collab MAU** — M1RR feeds into it; improving M1RR is one of the fastest levers to grow Collab MAU.
- **M3RR / M6RR** — longer-horizon return rates to track whether month-1 returners keep compounding.
- **New user acquisition volume** — M1RR is a rate; pair it with raw cohort size to understand absolute impact.

### MAU Family — Active User Decomposition

The MAU family metrics add up to a clean identity, which makes them powerful for diagnosing growth:

$$\text{MAU} = \underbrace{\text{New MAU}}_{\text{acquisition}} + \underbrace{\text{Returning MAU}}_{\text{retention}}$$

$$\text{Returning MAU} = \underbrace{\text{Repeat MAU}}_{\text{steady repeaters}} + \underbrace{\text{Reactivated MAU}}_{\text{comeback users}}$$

This decomposition lets you pinpoint *where* MAU is growing or shrinking without needing to look at raw cohort tables.

---
#### MAU — Monthly Active Users

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Lagging indicator |
| **Cadence** | Daily (rolling 28-day window) |
| **Formula** | $\text{MAU} = \lvert\{u : u \text{ active in trailing 28d}\}\rvert$ |
| **Target / Threshold** | _(to be defined)_ |
| **Data Source** | _(platform event log)_ |

**What it tells you:** The total distinct active user count over the most recent 28-day window — the standard baseline volume metric.

**Key distinction:** The 28-day rolling window (not a calendar month) keeps the number stable across months of different lengths and makes it refreshable daily.

---
#### WAU — Weekly Active Users

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Leading indicator (relative to MAU) |
| **Cadence** | Daily (rolling 7-day window) |
| **Formula** | $\text{WAU} = \lvert\{u : u \text{ active in trailing 7d}\}\rvert$ |
| **Target / Threshold** | _(to be defined — often expressed as WAU/MAU stickiness ratio)_ |
| **Data Source** | _(platform event log)_ |

**What it tells you:** Short-horizon active user count. Because WAU reacts faster than MAU, it is a leading signal for MAU movement — a drop in WAU for two consecutive weeks will almost certainly depress the next MAU reading.

**Common companion metric:** $\text{Stickiness} = \text{WAU} / \text{MAU}$ — a high ratio (e.g. > 0.5) means users are returning multiple times per month.

---
#### New MAU

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Leading indicator |
| **Cadence** | Daily (rolling 28-day window) |
| **Formula** | $\text{New MAU} = \lvert\{u \in \text{MAU} : \text{first\_active}(u) \in \text{current 28d}\}\rvert$ |
| **Target / Threshold** | _(to be defined)_ |
| **Data Source** | _(platform event log + user registration log)_ |

**What it tells you:** How many of today's MAU are brand new to the platform — the acquisition pulse.

**Key distinction:** "New" means first active use falls within the *current* 28-day window, not first registration. A user who signed up six months ago but never returned until this window is counted as New MAU.

---
#### Returning MAU

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Lagging indicator |
| **Cadence** | Daily (rolling 28-day window) |
| **Formula** | $\text{Returning MAU} = \text{MAU} - \text{New MAU}$ |
| **Target / Threshold** | _(to be defined — higher share of Returning vs. New MAU signals a maturing, sticky product)_ |
| **Data Source** | _(derived from MAU and New MAU)_ |

**What it tells you:** The retention-driven portion of MAU — users who had *some* prior history with the platform before the current window.

**Important:** Returning MAU includes both steady repeaters (Repeat MAU) *and* users who were dormant for a period (Reactivated MAU). Do not interpret it as "active every month."

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Lagging indicator |
| **Cadence** | Monthly |
| **Formula** | $\text{Reactivated MAU} = \text{Returning MAU} - \text{Repeat MAU}$ |

**What it tells you:** Users who were absent for at least one full 28-day window and have now come back. Separating this from Repeat MAU matters: a product relying heavily on reactivation rather than repeat usage has a retention problem, even if MAU looks healthy.

| Field | Detail |
|-------|--------|
| **Category** | Adoption & Reach |
| **Type** | Leading indicator |
| **Cadence** | Daily / weekly / monthly |
| **Formula** | $\lvert\{v : v \text{ visited in period}\}\rvert$ — browser/device-specific (not user-specific) |
| **Target / Threshold** | _(to be defined)_ |
| **Data Source** | _(Adobe Analytics — browser cookie-based)_ |

**What it tells you:** Pre-authentication reach — how many distinct browsers touched the product in the period. Because it is browser-based (not authenticated user-based), it can overcount users on multiple devices and undercount shared devices.

**Key distinction from MAU:** Unique Visitors includes unauthenticated traffic; MAU counts only authenticated or identified users. Unique Visitors ≥ MAU is typical.

### Conversion, Monetization & Retention Metrics

These metrics track the funnel from free usage → trial → paid subscription, and the leakage from it via churn. They form a connected system:

$$\text{Free MAU} \xrightarrow{\text{Free→Trial Rate}} \text{Trial Starts} \xrightarrow{\text{Trial→Paid Rate}} \text{Gross New Units} \xrightarrow{} \text{GNARR}$$
$$\text{Active Subscribers} \xrightarrow{\text{Churn Rate}} \text{Net Cancellation Units}$$

---
#### Gross New Units (GNU)

| Field | Detail |
|-------|--------|
| **Category** | Conversion, Monetization & Retention |
| **Type** | Lagging indicator |
| **Cadence** | Monthly / quarterly |
| **Formula** | $\text{GNU} = \text{Gross Purchase Units} - \text{Return Units}$ |
| **Target / Threshold** | _(to be defined)_ |
| **Data Source** | _(subscription management system)_ |

**What it tells you:** Net new subscription units added before accounting for ongoing churn or seat reductions. It is the unit-based counterpart to GNARR. "Gross" means walk effects (upgrades, downgrades, expansions) are excluded — this isolates pure new acquisition.

---
#### GNARR — Gross New Annual Recurring Revenue

| Field | Detail |
|-------|--------|
| **Category** | Conversion, Monetization & Retention |
| **Type** | Lagging indicator |
| **Cadence** | Monthly / quarterly |
| **Formula** | $\text{GNARR} = \text{Gross Purchase ARR} - \text{Returns ARR}$ |
| **Extended formula** | $\text{GNARR} = \text{First Purchase ARR} + \text{Resubscription ARR} + \text{Additional Purchase ARR} - \text{Returns ARR}$ |
| **Target / Threshold** | _(to be defined)_ |
| **Data Source** | _(finance / subscription management system)_ |

**What it tells you:** The revenue equivalent of GNU — how much new annualized recurring revenue was added in the period, before any churn or contraction offsets. The extended formula breaks it into components that reveal *how* new ARR is being generated (truly new customers vs. win-backs vs. expansion).

**Key distinction from Net New ARR:** GNARR is gross — it does not subtract churned ARR. Pair it with Churn ARR to get to Net New ARR.

---
#### Churn Rate

| Field | Detail |
|-------|--------|
| **Category** | Conversion, Monetization & Retention |
| **Type** | Lagging indicator |
| **Cadence** | Monthly |
| **Formula** | $\text{Churn Rate} = \dfrac{\text{Net Cancellation Units}}{\text{Beginning Units} + \text{Gross New Units}}$ |
| **Target / Threshold** | _(to be defined — also track Cancellation ARR / Units in parallel)_ |
| **Data Source** | _(subscription management system)_ |

**What it tells you:** The proportion of the subscriber base that cancelled in the period. The denominator — Beginning Units + GNU — adjusts for units added during the period, giving a fair rate against the full eligible base rather than just the opening count.

**Why it matters alongside GNARR:** GNARR and Churn Rate together determine net ARR growth. Strong GNARR with high churn is a treadmill — the CoE must watch both simultaneously.

**Related metric:** Churn ARR (revenue lost to cancellations) pairs with GNARR to compute Net New ARR.
