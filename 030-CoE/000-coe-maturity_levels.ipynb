# Center of Excellence (CoE)

## Data Maturity Model

| **Dimension**      | **1 — Foundational** | **2 — Emerging** | **3 — Mature** | **4 — Innovative** |
|---------------------|----------------------|------------------|----------------|--------------------|
| **Data Foundation** | Siloed, limited access. Questionable veracity. | Org invests in better data access. Key metrics are “certified”. | Defined owners and data stewards. Metrics and dashboard maintenance. Robust data infra for most data. | Cutting-edge data infrastructure allows for computationally intensive DS work (predictive modeling / ML / LLMs). |
| **Data Culture** | Some employees understand the value of data. No or limited data-driven experimentation / launch culture. Analysts focused on reporting and dashboards. | Some teams integrate data-driven processes and decision-making around experimentation and launch. Analysts start to work on DS-related areas (predictions, forecasting, experiment design). | Key organizational motions (Planning, CA/PPC, Product Launches, Marketing, etc.) centered around data products owned by DS/DE teams. | Org leans on data to help innovate for the future. New and innovative data analysis based on the latest science (e.g., applying advances in agentic systems for platform use cases). |
| **Data Tooling** | Ad-hoc analyst engagement. Mostly reactive. Stakeholder-initiated analysis of specific questions. | Formalized engagements with embedded analysis. Certain reporting and analysis are operationalized (OKRs, Cost, etc.). | Analysis in several areas is streamlined and well-defined. Specialized horizontal DS teams (OKR DS, Forecasting DS, Experiments DS). DS team is key for operational decision-making. | DS team is considered essential to both strategic planning and operational decision-making. |
| **Data Activation** | Reporting on key metrics. Deep dives in anomalies on request. | Operational analytics — KPIs reported in a set cadence. Planned deep dives and experiments to uncover insights in specific areas. | New insights and experimentation ideas to drive customer retention and revenue. | Identifying feature gaps and latent problems (e.g., surfacing customer pain points, predicting product usage, etc.). |


# Vision

| ----------------------- | -------------------------------------------------------------------------- | ------------------------------------------------------ |

# Quick Wins (first 60–90 days)
| Focus Area          | Concrete Step                                                                    | Deliverable                                         | Impact                        |
| ------------------- | -------------------------------------------------------------------------------- | --------------------------------------------------- | ----------------------------- |
| **Data Foundation** | Identify & certify 3 key datasets (e.g., users, sessions, revenue)               | “Certified Tables” badge in Databricks + short wiki | Visible quality improvement   |
| **Data Culture**    | Host 2 “Data Demo” sessions — show insights from certified data                  | Short presentation + recording                      | Builds excitement, visibility |
| **Data Tooling**    | Create 2–3 reusable Databricks notebooks (e.g., experiment analysis, KPI trends) | Shared repo + documentation                         | Promotes reusability          |
| **Data Activation** | Run one pilot with a PM (e.g., feature adoption funnel)                          | Mini case study                                     | Demonstrates applied value    |


# Kickoff: Understand Starting Point
| Area                | Typical Situation in a Platform Team                                                                 | Quick Assessment Method                                                              |
| ------------------- | ---------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------ |
| **Data Foundation** | Multiple product teams own data differently, no consistent ownership, limited certified data sources | **Interview** 3–5 key teams → ask where their “source of truth” is for key metrics       |
| **Data Culture**    | Reporting ≠ decisions. PMs/Designers rarely test hypotheses with data                                | Run a short **survey**: *“How often do you use data to inform your work?”*               |
| **Data Tooling**    | Reactive data requests, low reproducibility, dashboards built ad-hoc                                 | **Inventory**: dashboards, notebooks, scripts, owners                                    |
| **Data Activation** | Most insights = descriptive (what happened)                                                          | **Review** last 5 insights: were they descriptive, diagnostic, predictive, prescriptive? |
